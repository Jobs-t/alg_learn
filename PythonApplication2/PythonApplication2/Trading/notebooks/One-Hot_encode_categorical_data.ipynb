## Observations:

We have the following categorical data:
    1. Security_ID --> 18, 73, 102. Although the information is given in numbers, it still specifies a category.
    2. initiator --> Seller('S') or buyer('B')
    3. transtype< number > --> Trade('T') or Quote('Q')

## Next:

Drop the row_id column. Convert other columns into one-hot encoding form

```python
#Convert security_id into one-hot encoding
index = train_set.columns.get_loc('security_id')
train_set.insert(index+1, 'is_security_18', 0)
train_set['is_security_18'] = train_set['security_id'].apply(lambda x: 1 if x == 18 else 0)
train_set.head()
```

```python
index = train_set.columns.get_loc('security_id')
train_set.insert(index+1, 'is_security_102', 0)
train_set['is_security_102'] = train_set['security_id'].apply(lambda x: 1 if x == 102 else 0)
train_set.head()
```

```python
index = train_set.columns.get_loc('security_id')
train_set.insert(index+1, 'is_security_73', 0)
train_set['is_security_73'] = train_set['security_id'].apply(lambda x: 1 if x == 73 else 0)
train_set.head()
```

```python
#And ultimately drop the initial column itself
train_set = train_set.drop('security_id', 1)
train_set.head()
```

```python
#One-Hot encode initiator
index = train_set.columns.get_loc('initiator')
train_set.insert(index+1, 'is_seller', 0)
train_set.insert(index+1, 'is_buyer', 0)
train_set['is_seller'] = train_set['initiator'].apply(lambda x: 1 if x == 'S' else 0)
train_set['is_buyer'] = train_set['initiator'].apply(lambda x: 1 if x == 'B' else 0)
train_set.head()
```

```python
#And ultimately drop the initial column itself
train_set = train_set.drop('initiator', 1)
train_set.head()
```

```python
#Now we need to do the same for every transtype
for i in range(1, 51):
    column_name = 'transtype' + str(i)
    new_column_name_Q = 'transtype' + str(i) + '_is_Q'
    new_column_name_T = 'transtype' + str(i) + '_is_T'
    
    index = train_set.columns.get_loc(column_name)
    train_set.insert(index+1, new_column_name_Q, 0)
    train_set.insert(index+1, new_column_name_T, 0)
    
    train_set[new_column_name_Q] = train_set[column_name].apply(lambda x: +1 if x == 'Q' else 0)
    train_set[new_column_name_T] = train_set[column_name].apply(lambda x: +1 if x == 'T' else 0)
    
    #drop the initial column itself
    train_set = train_set.drop(column_name, 1)
```
